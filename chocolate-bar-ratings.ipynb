```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# Rename Column

```python
df.columns
```

```python
columns = list(df.columns)
re_columns = []
for n in columns:
    re_columns.append(n.replace('\n','_').replace(' ','_'))
re_columns
```

```python
re_columns[0] = "Company"
df = df.rename(columns=dict(zip(df.columns,re_columns)))
df.columns
```

# Drop null values

```python
df.isnull().sum()
```

```python
df[df.isnull().any(axis=1)]
```

```python
df = df.dropna()
df.isnull().sum()
```

```python
df['Bean_Type'].value_counts().head()
list(df['Bean_Type'][0:10])
```

# Replace Bean_Type to None

```python
def rep_space(x):
    if(x in '\xa0'):
        return 'None'

df['Bean_Type'] = df['Bean_Type'].apply(rep_space)
df.head()
```

# Convert Cocoa_Percent to numerical values

```python
df['Cocoa_Percent'] = df['Cocoa_Percent'].str.replace('%','').astype(float)/100
df.head()
```

**Question to answer**

1. which company get the highest rating?
2. what coaco percantage make the best chocolate?
3. Does cocoa percentage correlate with rating?
4. Trend of chocolate ratings over time?

# what coaco percantage make the best chocolate?

```python
#Average rating per cocoa percentage
cocoa_rating = df.groupby('Cocoa_Percent')['Rating'].mean()

plt.figure(figsize=(10,6))
plt.plot(cocoa_rating.index,cocoa_rating.values)
plt.xlabel("Cocoa Percentage")
plt.ylabel("Average Rating")
plt.title("Average rating by cocoa percentage")
plt.show()
```

# Does cocoa percentage correlate with rating?

```python
# Correlation
corr = df['Cocoa_Percent'].corr(df['Rating'])
print("correlation between Cocoa Percentage and Rating : ",corr)

#Visualize
plt.figure(figsize=(15,6))
sns.scatterplot(x = 'Cocoa_Percent', y = 'Rating', data = df)
plt.xlabel("Cocoa Percentage")
plt.ylabel("Average Rating")
plt.title("cocoa percentage VS rating")
plt.show()
```

- This show that rating does not depends on cocoa percentage

```python
from scipy.stats import pearsonr

corr, p_value = pearsonr(df['Cocoa_Percent'], df['Rating'])

print(f"Correlation: {corr:.3f}")
print(f"P-value: {p_value:.4f}")
```

# Trend of chocolate ratings over time?

```python
df['Year'] = pd.to_numeric(df['Review_Date'], errors='coerce')

year_rating = df.groupby('Review_Date')['Rating'].mean()

plt.figure(figsize=(15,6))
plt.plot(year_rating.index,year_rating.values)
plt.xlabel("Review_Date")
plt.ylabel("Rating")
plt.title("chocolate ratings over time")
plt.show()
```

# Trend of Cocoa percentage over time

```python
df['Year'] = pd.to_numeric(df['Review_Date'], errors='coerce')

year_rating = df.groupby('Year')['Cocoa_Percent'].mean()

plt.figure(figsize=(15,6))
plt.plot(year_rating.index,year_rating.values)
plt.xlabel("Year")
plt.ylabel("Cocoa Percentage")
plt.title("Cocoa Percentage over time")
plt.show()
```

- In 2008, the average chocolate rating was lower than in other years, despite having the highest average cocoa percentage. But after 2010, rating has increasing trend till 2016 (present time).
- In 2009, the average cocoa percentage dropped to its lowest point across the dataset, while the average rating increased compared to 2008.
- Overall, the trend in ratings appears to increase slightly for chocolates with cocoa percentages in the range of 70.5% to 72.2%.

# Which company gets the highest rating?

```python
#Average rating per company
company_rating = df.groupby('Company')['Rating'].mean().sort_values(ascending = False)

#Top 10 highest rating 
top_company = company_rating[company_rating >= 3.75]
print(top_company)

top_companies = top_company.index.tolist()

#Visualize
plt.figure(figsize=(10,6))
sns.barplot(x = top_company.values, y = top_company.index, palette="viridis")
plt.xlabel("Average Rating")
plt.title("Top Company (Rating >= 3.75) by average rating")
plt.show()
```

# Do the top companies have typical cocoa percentage ranges?

```python
top_df = df[df['Company'].isin(top_companies)]

#Average rating per company
company_cocoa = top_df.groupby('Company')['Cocoa_Percent'].mean().sort_values(ascending = False)

#Visualize
plt.figure(figsize=(10,6))
sns.barplot(x = company_cocoa.values, y = company_cocoa.index, palette="viridis")
plt.xlabel("Average Cocoa")
plt.title("Top Company (Rating >= 3.75) by average cocoa Percentage")
plt.show()

average_cocoa_percent = company_cocoa.mean()
print("Average Cocoa Percentage of Companies that have rating more than 3.75 is {:.2f}%".format(average_cocoa_percent*100))
```

Companies with average ratings ≥ 3.75 tend to produce chocolates with an average cocoa percentage of approximately 70%, suggesting a preference for moderately high cocoa content.

# Conclusion

This analysis examined the relationship between chocolate ratings, cocoa percentage, and time to determine whether cocoa content is a key driver of chocolate quality.

Overall, the results show that Cocoa percentage is not a strong driver of chocolate ratings. In 2008, chocolates had the highest average cocoa percentage but received lower average ratings compared to other years. In contrast, in 2009, the average cocoa percentage decreased to its lowest point in the dataset, while the average chocolate rating increased. After 2010, average chocolate ratings exhibited an upward trend, while cocoa percentages remained relatively stable. A slight increase in ratings was observed for chocolates with cocoa percentages in the range of 70.5% to 72.2%.

In addition, companies with average ratings of 3.75 or higher produced chocolates with an average cocoa percentage of 69.79%, indicating a preference for moderately high cocoa content among top-rated producers.

These findings suggest that while cocoa percentage contributes to chocolate characteristics, other factors—such as ingredient balance, sweetness, and production quality—likely play a more significant role in determining overall ratings.

Therefore, manufacturers aiming to achieve higher ratings should consider cocoa percentage as one component of their production strategy rather than the sole determinant of chocolate quality.
